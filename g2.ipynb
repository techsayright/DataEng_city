```python
import pandas as pd

url1 = "/home/darshu/Documents/python projects/DataEngineering/North&East_House_Rent.xml"
url2 = "/home/darshu/Documents/python projects/DataEngineering/South_House_Rent.json"
url3 = "/home/darshu/Documents/python projects/DataEngineering/West_House_Rent.csv"

df1 = pd.read_xml(url1)
df2 = pd.read_json(url2)
df3 = pd.read_csv(url3)

dataframe = pd.concat([df1, df2, df3])
dataframe
```

```python
dataframe.isnull().sum()
```

```python
dataframe = dataframe.dropna()
dataframe
```

```python
dataframe = dataframe.rename(columns = {'locality': 'location'})
dataframe
```

```python
dataframe.drop(dataframe[(dataframe['bathroom'] != '2 bathrooms') & (dataframe['bathroom'] != '1 bathrooms') & (dataframe['bathroom'] != '3 bathrooms') & (dataframe['bathroom'] != '4 bathrooms') & (dataframe['bathroom'] != '7 bathrooms') & (dataframe['bathroom'] != '5 bathrooms') & (dataframe['bathroom'] != '6 bathrooms') & (dataframe['bathroom'] != '9 bathrooms') & (dataframe['bathroom'] != '10 bathrooms')].index, inplace=True)
```
